Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
- high - длинный(от 42)
- middle - средний(от 35 до 42)
- low - маленький(до 35)

```python
pg.loc[pg['bill_length_mm'] > 42, 'new_col'] = 'high'
```

```python
pg.loc[(pg['bill_length_mm'] > 35) & (pg['bill_length_mm'] <= 42), 'new_col'] = 'middle'
```

```python
pg.loc[pg['bill_length_mm'] <= 35, 'new_col'] = 'low'
```

Изобразить гистограмму по flipper_length_mm с оттенком new_col. Сделать анализ


```python
sns.histplot(data=pg, x='flipper_length_mm', hue = 'new_col')
```
